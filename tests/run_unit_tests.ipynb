```python
# Adds the parent directory of the current working directory to the Python search path
import sys 
import os

# Get the current working directory
current_path = os.getcwd()

# Get the src folder path by navigating one level up from the current path and then going into src
src_path = os.path.join(os.path.dirname(current_path), "src")

# Add the root folder path to the system path
sys.path.append(src_path)

# Print a message confirming the root folder added to sys.path
print(f"Added the following path: {src_path}")
```
